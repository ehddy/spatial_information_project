```python
class Groupby_commodity():
        """
        0.1 초기 설정 
        """
        def __init__(self, total_data, total_data_domestic, name_data, year):
            
            self.total_data = total_data 
            self.total_data_domestic = total_data_domestic
            
            
            for i in range(4):
                self.total_data[i] = self.total_data[i].iloc[:, 1:].rename(columns={'Unnamed: 1':'상품분류'}).set_index('상품분류')
                self.total_data_domestic[i] = self.total_data_domestic[i].iloc[:, 1:].rename(columns={'Unnamed: 1':'상품분류'}).set_index('상품분류')
#                 data['대분류'] = list(pd.Series(data.index.values).apply(lambda x: dic_pre(x, dic)))

#                 data = data.replace(0, np.NaN)
            
            self.name_data = name_data 
            self.year = year 
        
        
            self.total_group_count = []
            self.group_name_list = []
            self.total_name_or_idx_list = []
            self.matching_dic = {}
            
            
             
            
        def standard_choise(self):   
            print('===================================')
            print('원하시는 분류 기준을 선택해주세요. ')
            print('1. 기본분류 \n 2. 소분류 \n 3. 중분류 \n 4. 대분류')
            choise = int(input('번호 선택 : '))
            print('분류 기준을 설정하였습니다.')
            print('===================================')
    
            if choise == 1: 
                self.name_data = self.name_data.iloc[:381, [0, 1]].dropna(how='any').reset_index(drop=True)
                self.group_name_len = len(self.name_data)
            elif choise == 2:
                self.name_data = self.name_data.iloc[:381, [2, 3]].dropna(how='any').reset_index(drop=True)
                self.name_data.columns = ['코드', '부문명']
                self.group_name_len = len(self.name_data)
            elif choise == 3:
                self.name_data = self.name_data.iloc[:381, [4, 5]].dropna(how='any').reset_index(drop=True)
                self.name_data.columns = ['코드', '부문명']
                self.group_name_len = len(self.name_data)
            else:
                self.name_data = self.name_data.iloc[:381, [6, 7]].dropna(how='any').reset_index(drop=True)
                self.name_data.columns = ['코드', '부문명']
                self.group_name_len = len(self.name_data)
                

                
        def name_select(self):
            display(self.name_data)
        
            group_count = int(input('해당 상품 분류들로 몇 개의 그룹을 생성하고 싶습니까?(최종 산업 개수): '))
            print()
            self.total_group_count.append(group_count)
            print('=== 그룹 개수 입력 완료 ===')
            print('======================================================')
            for i in range(1, group_count+1):
                name = input('{}번 그룹의 이름을 입력해주세요 : '.format(i))
                self.group_name_list.append(name)
            print('===', group_count,'개의 그룹 이름 입력 완료 ===')
            print(self.group_name_list)
            print('======================================================')
            
            
                            
            
            
            
            
        def name_matching(self):
            print('최종 그룹 목록: {}'.format(self.group_name_list))
            print('======================================================')
            print('1. 각 그룹에 포함될 상품 분류의 인덱스 지정 \n 2. 각 그룹에 포함될 상품 분류의 이름 지정')
            
            index_or_name = int(input('어떤 방식으로 각각의 그룹에 포함될 상품 분류를 구성하시겠습니까?(번호 선택) : '))
            print('======================================================')
            
            # 중복값 찾기용 list
            overlap_list = []
            
            total_len = self.group_name_len
            total_list = []          
            for name in self.group_name_list:              
                name_or_idx_list = []
                if total_len == 0:   
                    break 
                if index_or_name == 1:
                    while True:                           
                        idx_range = input('{}에 포함시킬 상품 분류 인덱스 범위를 입력해주세요: ex) 1-{} '.format(name, self.group_name_len))                
                        try: 
                            idx_range = int(idx_range)
                            if idx_range in overlap_list or idx_range > self.group_name_len:
                                print()
                                print('범위를 초과했거나 이미 다른 그룹에 중복된 상품 분류가 포함되어 있습니다. 다시 입력해주세요. ')
                                print('사용된 상품분류 index 리스트: ')
                                print(overlap_list)
                                continue
                            total_len -= 1
                            name_or_idx_list.append(idx_range)
                            overlap_list.append(idx_range)
                        except:
                            idx_range_list = idx_range.split('-')
                            start_num = int(idx_range_list[0])
                            
                            final_num = int(idx_range_list[1])
                            
                            idx_range = [i for i in range(start_num, final_num+1)]
                            
                            overap = 0
                            
                            for idx in idx_range:
                                if idx in overlap_list or idx > self.group_name_len:
                                    print()
                                    print('범위를 초과했거나 이미 다른 그룹에 중복된 상품 분류가 포함되어 있습니다. 다시 입력해주세요. ')
                                    print('사용된 상품분류 index 리스트: ')
                                    print(overlap_list)
                                    overap = 1
                                    break
                            if overap != 0:
                                continue
                                        
                            name_or_idx_list.extend(idx_range)
                            overlap_list.extend(idx_range)
               
                            total_len -= final_num - start_num + 1 
                        
                        if total_len == 0:   
                            print()
                            total_list.append(name_or_idx_list)
                            print('0개의 그룹이 남았습니다. 분류 선택을 종료합니다.')
                            print('================================================')
                            break        
                        
                        next_y_n = input('{}개의 상품 분류 그룹이 남았습니다. 다음 산업으로 넘어가겠습니까?(yes or no) '.format(total_len))
                        print()
                        if next_y_n == 'yes':
                            if name == self.group_name_list[-1]:
                                print('마지막 그룹입니다. 나머지 상품 분류 인덱스를 추가해주세요')
                                continue
                            total_list.append(name_or_idx_list)
                            break
                        else:
                            continue    
                else:
                    while True:                  
                        c_name = input('{}에 포함시킬 상품 분류 이름을 입력해주세요: '.format(name))
                        if c_name in overlap_list or c_name not in list(self.name_data['부문명']):
                                print('이미 다른 그룹에 중복된 상품 분류가 포함되어 있거나 존재하지 않는 상품 분류입니다. 다시 입력해주세요. ')
                                print('사용된 상품분류 이름 리스트: ')
                                print(overlap_list)
                                continue
                        total_len -= 1               
                        overlap_list.append(c_name)
                        name_or_idx_list.append(c_name)                                 
                        if total_len == 0:   
                            print()
                            total_list.append(name_or_idx_list)
                            print('0개의 그룹이 남았습니다. 분류 선택을 종료합니다.')
                            print('================================================')
                            break                    
                        next_y_n = input('{}개의 상품 분류 그룹이 남았습니다. 다음 산업으로 넘어가겠습니까?(yes or no) '.format(total_len))        
                        print()
                        if next_y_n == 'yes':
                            if name == self.group_name_list[-1]:
                                print('마지막 그룹입니다. 나머지 상품 분류 인덱스를 추가해주세요')
                                continue
                        
                            total_list.append(name_or_idx_list)
                            break
                        else:
                    
                            continue
                                  
            self.total_name_or_idx_list.extend(total_list)
            dic = {}    
            for i, name in enumerate(self.group_name_list):
                dic[name] = self.total_name_or_idx_list[i]
                
            
            # 딕셔너리 뒤집기      
            for i, name in enumerate(self.group_name_list):
    
                for small_name in list(dic.values())[i]:

                    self.matching_dic[small_name] = name 
            
            # 만약 Index로 받아온 경우 인덱스를 상품명으로 바꿔서 딕셔너리를 형성해야함 
            if index_or_name == 1:
                total_idx = list(self.matching_dic.keys())
                
                for i in total_idx:
                    new_key = self.name_data.iloc[i-1][1]
                    self.matching_dic[new_key] = self.matching_dic.pop(i)

            print('분류 매칭을 완료했습니다.')
            print(self.matching_dic)
            print('=============================')
        
        
        
        # 재분류 열 추가 생성 함수(대분류 수준)
        def dic_pre(self, x, dic):
            try:
                return dic[x]
            except:
                return 0
        
        
        
    
        def groupping(self):
            
            # 정렬용 리스트 

            reindex_index = self.group_name_list.copy()
            
            
            reindex_index.extend(list(self.total_data[0].index)[self.group_name_len:])
            
            
            
            reindex_column = self.group_name_list.copy()
            reindex_column.extend(list(self.total_data[0].columns)[self.group_name_len:])
                
            
            
            groupping_list = []
            y = self.year
            metrix = self.total_group_count
            print('=======================================')
            print('그룹화 데이터 생성 프로그램 시작')
            print('=======================================')
        
            for data in self.total_data:
                data['그룹 분류'] = list(pd.Series(data.index.values).apply(lambda x: self.dic_pre(x, self.matching_dic)))

                data = data.replace(0, np.NaN)
                metrix_a_1 = data.loc[:'중간투입계', :].groupby('그룹 분류').sum()
                
                metrix_b_1 = data.loc['중간투입계':, :].drop('그룹 분류', axis=1)
                
            

                metrix_total_1 = pd.concat([metrix_a_1, metrix_b_1], axis=0).transpose()


                metrix_total_1['그룹 분류'] = list(pd.Series(metrix_total_1.index.values).apply(lambda x: self.dic_pre(x, self.matching_dic)))
    
                metrix_total_1 = metrix_total_1.replace(0, np.NaN)

                metrix_a_2 = metrix_total_1.loc[:'중간수요계', :].groupby('그룹 분류').sum()
                metrix_b_2 = metrix_total_1.loc['중간수요계':, :].drop('그룹 분류', axis=1)

                metrix_total_2 = pd.concat([metrix_a_2, metrix_b_2], axis=0).transpose()
                metrix_total_2 = metrix_total_2.reindex(index=reindex_index, columns=reindex_column)
                
                groupping_list.append(metrix_total_2)
                print('{}년 데이터 전처리 완료'.format(y))
                y += 1 
                
                
            print('=======================================')
            print('그룹화된 데이터 생성을 완료하였습니다.')
            print('=======================================')
                

            
            return groupping_list 


        
```
