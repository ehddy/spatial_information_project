```python
total_result_data = []
i = 0
for data in total_data:

    data.drop('Unnamed: 0', axis=1, inplace=True)
    
    if i >= 2:
        data.rename(columns={'Unnamed: 1':'구분', 'Unnamed: 37':'총계'},inplace=True)
    else:
        data.rename(columns={'Unnamed: 1':'구분', 'Unnamed: 35':'총계'},inplace=True)
    data['구분'] = data['구분'].fillna('총계')
    data.set_index('구분', inplace=True)
    data.fillna(0, inplace=True)
    data.replace(' ', 0, inplace=True)
    data = data.astype('float')
    
    total_result_data.append(data)
    i += 1 
```

```python
# 자동으로 횟수 정해주는 RAS 
def Ras_1(total_data, metrix_num):
    
    result_metrix_list = []
    idx = 0
    for data in total_data:
        
        if idx >= 2:
            metrix_num = 35

        
        num = 0 

        
        standard_sum = data.loc['총계', :'수입'].values

        change_sum = data.loc[:'수입', '총계'].values

        input_coeff = data.loc[:'수입', :'수입'].values / change_sum
            

        metrix = np.identity(metrix_num)
        r = np.zeros(metrix_num)
        s = np.zeros(metrix_num)
        
        # 1차 잠정거래행렬
        m = np.dot(input_coeff , standard_sum * metrix)
        print('제{}차 잠정거래행렬 : '.format(num+1))
        print(m, end='\n\n')
        

        while True:
            i = 0
            if num % 2 == 0: 
                w = np.array([sum(m[i]) for i in range(metrix_num)])
                before_r = r 
                r = standard_sum / w 
                r = r.astype('float64')
                m = np.dot(r * metrix, m)
                num += 1 




            elif num % 2 != 0:
                z = np.array([sum(m[:, i]) for i in range(metrix_num)])
                before_s = s 
                s = standard_sum / z
                s = s.astype('float64')
                m = np.dot(m, s * metrix)
                num += 1 
            if num == 999:
                break



        

        print('------------------------------------------', end='\n\n')
        print('제{}차 잠정거래행렬 : '.format(num+1))
        print(m, end='\n\n')
        
        result_data = pd.DataFrame(data=m, columns=list(data.columns)[:-1], index=list(data.index)[:-1])
        result_metrix_list.append(result_data)
        print('행렬 조정 완료')
        print('==================================')
        idx += 1 
    return result_metrix_list
```

```python
final_list = Ras_1(total_result_data, 33)
```
